<h2 style="color:blue;">Import numpy, matplotlib and LinearRegression from sklearn</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

<h2 style="color:blue;">Define the Sample Data</h2>

```python

# Sample data
# X = size of the house (in square feet)
# y = price of the house (in Rs in Lakhs )
X = np.array([[300], [400], [500], [600], [700]])
print(X)
y = np.array([150, 200, 250, 300, 350])
print(y)

```

```python

# Create a linear regression model
model = LinearRegression()
```

```python

# Train the model
model.fit(X, y)
```

```python

# Predict
predicted = model.predict(X)
```

```python

# Print model parameters
print("Slope (Coefficient):", model.coef_)
print("Intercept:", model.intercept_)
```

```python

# Plotting
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, predicted, color='red', linewidth=2, label='Regression line')
plt.xlabel("Size (sq ft)")
plt.ylabel("Price (in Lakhs)")
plt.title("House Price Prediction")
plt.legend()
plt.grid(True)
plt.show()
```

```python
X_test = np.array([[320],[455]])
# Predict
predicted = model.predict(X_test)
print(predicted)
```

```python
plt.plot(X, y, color='red', linewidth=2, label='Regression line')
plt.scatter(X, y, color='blue', label='Actual data')
plt.scatter(X_test, predicted, color='green', label='test data')

plt.xlabel("Size (sq ft)")
plt.ylabel("Price (in Lakhs)")
plt.title("House Price Prediction")
plt.legend()
plt.grid(True)
plt.show()
```
