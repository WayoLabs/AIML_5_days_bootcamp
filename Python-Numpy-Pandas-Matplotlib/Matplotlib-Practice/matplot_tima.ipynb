```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =[1,2,3,4,5]
y=[5,6,7,8,5]

plt.plot(x,y)

plt.show()
```

```python
# Generate x values
x = np.linspace(-10, 10, 100)  # 100 points from -10 to 10

# Define the line equation y = mx + c
m = 2
c = 1
y = m * x + c

# Plot the line
plt.plot(x, y, label="y = 2x + 1", color="blue")

# Add labels and grid
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line Plot from Equation")
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
categories = ['Python', 'Java', 'C++', 'JavaScript']
popularity = [85, 80, 70, 75]

plt.bar(categories, popularity, color='skyblue')
plt.title("Programming Language Popularity")
plt.ylabel("Popularity (%)")
plt.show()
```

```python
slices = [50, 20, 30]
labels = ['AI', 'Web Dev', 'Data Science']
colors = ['#ff9999','#66b3ff','#99ff99']

plt.pie(slices, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Interest Distribution")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 9]

plt.scatter(x, y, color='red')
plt.title("Scatter Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot([1, 2, 3], [1, 4, 9])
axs[0, 1].bar(['A', 'B', 'C'], [3, 7, 2])
axs[1, 0].scatter([1, 2, 3], [3, 1, 4])
axs[1, 1].pie([10, 20, 70], labels=['X', 'Y', 'Z'])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Sample data
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03'],
    'Cases': [1000, 1500, 1800]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

plt.plot(df['Date'], df['Cases'], marker='o')
plt.title("COVID-19 Daily Cases")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
data = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]
plt.hist(data, bins=12)
plt.show()
```

```python
x = np.logspace(0.1, 2, 100)  # values from ~1.26 to 100
y = np.log10(x)               # log base 10 of x

plt.plot(x, y)
plt.xscale('log')             # 🔥 Make X-axis logarithmic
plt.title("Logarithmic X-Axis Example")
plt.xlabel("Log-scaled X")
plt.ylabel("log10(x)")
plt.grid(True)
plt.show()
```
